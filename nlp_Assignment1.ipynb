```python
f = open("INDIAN_CONSTITUTION.pdf", "rb")
pdf_text = []
pdf_reader = PyPDF2.PdfReader(f)
for page in range(len(pdf_reader.pages)):
  page = pdf_reader.pages[page]
  pdf_text.append(page.extract_text())
f.close()
```

```python
for page in pdf_text:
  print(page)
  print("\n")
  print("\n")
  print("\n")
```
